## Testing Various Models

```python
label_Enconder = LabelEncoder()
```

```python
Credit_card_new["Attrition_Flag"]= label_Enconder.fit_transform(Credit_card_new["Attrition_Flag"])
Credit_card_new["Gender"] = label_Enconder.fit_transform(Credit_card_new["Gender"])
Credit_card_new["Education_Level"] = label_Enconder.fit_transform(Credit_card_new["Education_Level"])
Credit_card_new["Marital_Status"] = label_Enconder.fit_transform( Credit_card_new["Marital_Status"])
Credit_card_new["Income_Category"] = label_Enconder.fit_transform(Credit_card_new["Income_Category"])
Credit_card_new["Card_Category"] = label_Enconder.fit_transform(Credit_card_new["Card_Category"])
```

```python
Credit_card_new.head(10)
```

```python
X = Credit_card_new.drop(columns="Attrition_Flag")
y = Credit_card_new["Attrition_Flag"]
```

### KNN Classifier

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=1)
```

```python
model =KNeighborsClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
y_predict = model.predict(X_test)
```

```python
print(accuracy_score(y_test,y_predict))
```

```python
pd.crosstab(y_test,y_predict)
```

### Decision Tree

```python
model_tree = tree.DecisionTreeClassifier()
```

```python
model_tree.fit(X_train,y_train)
```

```python
y_tree_predict = model_tree.predict(X_test)
```

```python
print(accuracy_score(y_test,y_tree_predict))
```

```python
pd.crosstab(y_test,y_tree_predict)
```
